```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

#  Excercise 1
Create a figure object called fig using plt.figure() 

Use add_axes to add an axis to the figure canvas at [0,0,1,1]

Plot (x,y) on that axes and set the labels and titles

 




```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```
